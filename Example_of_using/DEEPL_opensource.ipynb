```python
CODE="/content/drive/MyDrive/DeepEL_opensource/DeepEL/DeepEL_baseline/DeepEL_codes/Change_sentence/Blink_for_changed_sentence.py"
INPUT_FILE="/content/drive/MyDrive/DeepEL_opensource/data/formal_Experiment1/Results_Change_sentence/ace2004.json"
OUTPUT_FILE="ace2004_changed.json"
OUTPUT_DIR="/content/drive/MyDrive/DeepEL_opensource/data/formal_Experiment1/Result_BLINK_for_changed"
```

**Step4**

At this step, the code merges the entity list generated from the original sentence with the entity list generated from the rewritten sentence.

```python
CODE="/content/drive/MyDrive/DeepEL_opensource/DeepEL/DeepEL_baseline/DeepEL_codes/Merge_result/Merge.py"
INPUT_FILE="/content/drive/MyDrive/DeepEL_opensource/data/formal_Experiment1/Result_BLINK_for_changed/ace2004_changed.json"
SECOND_INPUT_FILE="/content/drive/MyDrive/DeepEL_opensource/data/formal_Experiment1/Result_BLINK_Result/ace2004_changed.json"
OUTPUT_DIR="/content/drive/MyDrive/DeepEL_opensource/data/formal_Experiment1/Result_BLINK_merged"
OUTPUT_FILE="ace2004_changed.json"
```

```python
!python3 "$CODE" \
    --input_file "$INPUT_FILE" \
    --second_input_file "$SECOND_INPUT_FILE" \
    --output_dir "$OUTPUT_DIR" \
    --output_file "$OUTPUT_FILE"
```
